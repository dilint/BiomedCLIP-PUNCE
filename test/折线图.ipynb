```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
df_1 = np.array([[78.8,77.2,78.4,78.1,78.2,77.5],
               [82.7,82.6,83.1,83.5,83.2,82.7],
               [84.9,85.2,85.7,86,85.7,85.6],
               [86.4,86.6,86.9,86.8,86.75,86.6]])

df_2 = np.array([[78.3,78.8,79.5,78.1,78.5],
               [83.7,83.65,83.86,83.48,83.9],
               [85.8,85.95,86.33,86.02,86.25],
               [86.84,86.86,87.03,86.8,87.1]])

df_3 = np.array([[76.31,76.32,78.07,75.87,76.95],
               [81.73,81.74,83.46,81.69,80.53],
               [85,85.05,86.03,83.73,82.24],
               [86.27,86.21,86.81,85.07,84.89]])
```

```python
config = {
"font.family":'serif',
"font.size": 15,
"mathtext.fontset":'stix',
'font.family': 'Times New Roman, SimSun'
            }
plt.rcParams.update(config)          

fig,ax=plt.subplots(1,3,figsize=(15,5),dpi=200)
fig.subplots_adjust(hspace=0, wspace=0)


colors= ['#C5E5F4', '#8FD6EA', '#00B2D9', '#447C9C']
labels = ['k=5', 'k10', 'k=25', 'k=50']
titles = ['Batch size', 'Number of batch tokens from $H_{i}$', 'Temperature']
for i in np.arange(4):
    ax[0].plot(np.arange(6), df_1[i], color='gray', marker='o', markerfacecolor=colors[i], linestyle='--', markersize=12, linewidth=0.5)
    ax[1].plot(np.arange(5), df_2[i], color='gray', marker='o', markerfacecolor=colors[i], linestyle='--', markersize=12, linewidth=0.5)
    ax[2].plot(np.arange(5), df_3[i], color='gray', marker='o', markerfacecolor=colors[i], linestyle='--', markersize=12, linewidth=0.5)

for i in np.arange(3):
    ax[i].set_ylim(74,88)
    ax[i].set_yticks([75.0, 77.5, 80.0, 82.5, 85.0, 87.5])
    ax[i].set_title(titles[i])
    ax[i].spines['top'].set_visible(False)  
    

ax[0].spines['right'].set_linewidth(1)  
ax[1].set_yticks([])
ax[1].spines['left'].set_visible(False)
ax[1].spines['right'].set_linewidth(0.5)  
ax[2].set_yticks([])
ax[2].spines['left'].set_visible(False)



lineticks=[75.0, 77.5, 80.0, 82.5, 85.0, 87.5]

for i in np.arange(3):
    for j in np.arange(6):
        ax[i].axhline(y=lineticks[j], color='gray', linestyle='-', linewidth=0.5)  

    
ax[0].axhline(y=75.0, color='gray', linestyle='-', linewidth=0.5)  

ax[0].set_xticks(np.arange(6))
ax[0].set_xticklabels(np.power(2, np.arange(4, 10)))
ax[0].set_xlim(-0.5,5.5)

ax[1].set_xticks(np.arange(5))
ax[1].set_xticklabels(np.power(2, np.arange(9, 14)))
ax[1].set_xlim(-0.5,4.5)

ax[2].set_xticks(np.arange(5))
ax[2].set_xticklabels([0.0005, 0.005, 0.01, 0.05, 0.01])
ax[2].set_xlim(-0.5,4.5)

for i in np.arange(4):
    ax[0].scatter([], [], c=colors[i], label=labels[i], s=200, marker='s')
    
ax[0].legend(frameon=True, labelspacing=1, bbox_to_anchor=(2.2, -0.1), ncol=4, fontsize=15)

plt.savefig('0614.png')
```

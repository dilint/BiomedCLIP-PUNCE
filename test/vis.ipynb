# ROC curve

```python

import numpy as np
import os
import matplotlib.pyplot as plt

config = {
"font.family":'serif',
"font.size": 12,
"mathtext.fontset":'stix',
'font.family': 'Times New Roman'}
plt.rcdefaults()
plt.rcParams.update(config)       

fig,ax = plt.subplots(1,1,figsize=(5,5),dpi=400)
fig.subplots_adjust(hspace=0, wspace=0)

roc_root_dir = 'roc_data'
feat_dirs = ['biomed3', 'biomed1', 'resnet1', 'resnet50']
legends = ['Ours', 'BiomedCLIP', 'ResNet50', 'ResNet50-MoCov2']
fprs = []
tprs = []
cmap_colors = plt.get_cmap('Paired').colors

colors = [cmap_colors[0], cmap_colors[2], cmap_colors[4], cmap_colors[6]]
font = {'fontsize': 20, 'fontfamily': 'Times New Roman'}

ax.set_xlim([0.0, 0.5])
ax.set_ylim([0.8, 1.0])
ax.set_yticks([0.85, 0.90, 0.95, 1.0])
ax.set_xlabel('False Positive Rate', fontdict=font)
ax.set_ylabel('True Positive Rate', fontdict=font)
ax.set_title('ROC curve', fontdict=font)

xticks = np.arange(0.1, 1.1, 0.1)
for i in np.arange(10):
    ax.axvline(x=xticks[i], color='#c1c1c1', linestyle='-', linewidth=0.7, zorder=0)
yticks = np.array([0.85, 0.90, 0.95, 1.0])
for i in np.arange(4):
    ax.axhline(y=yticks[i], color='#c1c1c1', linestyle='-', linewidth=0.7, zorder=0)

lw = 2
for i, feat_dir in enumerate(feat_dirs):
    fpr = np.load(os.path.join(roc_root_dir, feat_dir, 'fpr.npy'))
    tpr = np.load(os.path.join(roc_root_dir, feat_dir, 'tpr.npy'))
    auc = np.load(os.path.join(roc_root_dir, feat_dir, 'auc.npy'))
    plt.plot(fpr, tpr, color=colors[i],
            lw=4, label=f'{legends[i]} (auc = %0.4f)' % auc)
    
ax.plot([0, 1], [0, 1], color='black', lw=1.5, linestyle='--')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)  

ax.legend(loc="lower right", edgecolor='black', fontsize=12)
plt.savefig('output/roc_curve.png', bbox_inches='tight')
# plt.savefig('roc_curve.pdf', bbox_inches='tight')
```

# Vis for BiomedCLIP

```python
import torch
from PIL import Image
from torchvision import transforms
import os
from clip_visual_model import CLIPViT

device = torch.device('cuda:1') if torch.cuda.is_available() else torch.device('cpu')
model = CLIPViT().to(device)
model.load_from('biomedclip')
model.eval()
```

```python
import os
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

data_root = '/home/commonfile/TCTAnnotatedData/TCTAnnotated20210331/Annotated202103TCTFK-XIMEA-1200-YX-BZ-1'
wsi_name = '4867 M-G'
input_dir = os.path.join(data_root, wsi_name)
output_dir = 'output'

vis_patch = '14_9'
vis_image = vis_patch + '.jpg'
vis_txt = vis_patch + '.txt'
vis_image_path = os.path.join(input_dir, vis_image)
vis_txt_path = os.path.join(input_dir, vis_txt)
print(vis_txt_path)
boxes = []
with open(vis_txt_path, 'r') as file:
    lines = file.readlines()
    lines = lines[1:]
    for line in lines:
        n, x, y, w, h = line.strip().split('\t')
        boxes.append((int(n), int(x), int(y), int(w), int(h)))
print(boxes)

# 打开图片
box_image = Image.open(vis_image_path)
    
target_patch_size = (1024,1024)
precess = transforms.Compose([
    transforms.Resize(target_patch_size),
    transforms.ToTensor(),
])

img_path = vis_image_path
img = Image.open(img_path)
img = precess(img).unsqueeze(0).to(device)

with torch.no_grad():
    image_attention = model.get_last_selfattention(img)
    print(image_attention.shape)
    cls_attention = torch.mean(image_attention, dim=0)[0,1:].reshape(64,64)
    print(cls_attention.shape)
    
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 3, figsize=(10, 3), dpi=300)
fig.patch.set_facecolor('white')
fig.subplots_adjust(hspace=0, wspace=0.03)

box_image = Image.open(vis_image_path)
draw = ImageDraw.Draw(box_image)

ax[0].imshow(box_image.resize(target_patch_size))
ax[0].axis('off')

for box in boxes:
    n, x, y, w, h = box
    # rect = Rectangle((x, y), w, h, edgecolor='r', facecolor='red')  # 创建矩形对象
    # ax[0].add_patch(rect)  # 将矩形添加到子图中
    draw.rectangle((x, y, x+w, y+h), outline="red", width=20)

ax[1].imshow(box_image.resize(target_patch_size))
ax[1].axis('off')

ax[2].imshow(cls_attention.to('cpu'))
ax[2].axis('off')

plt.savefig(f'output/backbone_{vis_patch}.png', bbox_inches='tight')
```

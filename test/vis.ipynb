# ROC curve

```python

import numpy as np
import os
import matplotlib.pyplot as plt

config = {
"font.family":'serif',
"font.size": 12,
"mathtext.fontset":'stix',
'font.family': 'Times New Roman'}
plt.rcdefaults()
plt.rcParams.update(config)       

fig,ax = plt.subplots(1,1,figsize=(5,5),dpi=400)
fig.subplots_adjust(hspace=0, wspace=0)

roc_root_dir = 'roc_data'
feat_dirs = ['biomed3', 'biomed1', 'resnet1', 'resnet50']
legends = ['Ours', 'BiomedCLIP', 'ResNet50', 'ResNet50-MoCov2']
fprs = []
tprs = []
cmap_colors = plt.get_cmap('Paired').colors

colors = ['darkorange', 'lightgreen', 'blue', 'pink']
colors = ['#fbe5d6', '#e2f0d9', '#deebf7', '#fffff2']
colors = [cmap_colors[0], cmap_colors[2], cmap_colors[4], cmap_colors[6]]
font = {'fontsize': 20, 'fontfamily': 'Times New Roman'}

ax.set_xlim([0.0, 0.5])
ax.set_ylim([0.8, 1.0])
ax.set_yticks([0.85, 0.90, 0.95, 1.0])
ax.set_xlabel('False Positive Rate', fontdict=font)
ax.set_ylabel('True Positive Rate', fontdict=font)
ax.set_title('ROC curve', fontdict=font)

xticks = np.arange(0.1, 1.1, 0.1)
for i in np.arange(10):
    ax.axvline(x=xticks[i], color='#c1c1c1', linestyle='-', linewidth=0.7, zorder=0)
yticks = np.array([0.85, 0.90, 0.95, 1.0])
for i in np.arange(4):
    ax.axhline(y=yticks[i], color='#c1c1c1', linestyle='-', linewidth=0.7, zorder=0)

lw = 2
for i, feat_dir in enumerate(feat_dirs):
    fpr = np.load(os.path.join(roc_root_dir, feat_dir, 'fpr.npy'))
    tpr = np.load(os.path.join(roc_root_dir, feat_dir, 'tpr.npy'))
    auc = np.load(os.path.join(roc_root_dir, feat_dir, 'auc.npy'))
    plt.plot(fpr, tpr, color=colors[i],
            lw=lw, label=f'{legends[i]} (auc = %0.4f)' % auc)
    
ax.plot([0, 1], [0, 1], color='black', lw=1.5, linestyle='--')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)  

ax.legend(loc="lower right", edgecolor='black', fontsize=12)
plt.savefig('roc_curve.png', bbox_inches='tight')
```
